### mean

```python
import matplotlib.pyplot as plt
def mean(data):
  return sum(data)/len(data)

mean1 = mean([1,2,3,4]);
print("Mean :",mean1);
```

### varience

```python
def vari(data):
  value = ((i-mean(data))**2 for i in data)
  return sum(value)/len(data)


print("Varience :",vari([1,2,3,4,5]))
```

### Standard Deviation

```python
def std(data):
    varience = vari(data)
    return varience**0.5

print("Standard Deviation :",std([1,2,3,4,5,6]))
```

## Covarince

```python
def cov(data1, data2):
    n = len(data1)
    mean_data1 = mean(data1)
    mean_data2 = mean(data2)

    covariance = sum((data1[i] - mean_data1) * (data2[i] - mean_data2) for i in range(n)) / (n - 1)

    return covariance


data1 = [1, 2, 3, 4]
data2 = [7, 8, 9, 10]

covariance_result = cov(data1, data2)
print("Covariance:", covariance_result)
```

```python
def cor(data1,data2):
  covarience =cov(data1,data2)
  mean_data1 =mean(data1)
  mean_data2 =mean(data2)
  n = len(data1)

  prod_std1 = (sum((i-mean_data1)**2 for i in data1)/(n-1)) **0.5
  prod_std2 = (sum((i-mean_data2)**2 for i in data2)/(n-1)) **0.5

  correlation = covarience/(prod_std1*prod_std2)
  return correlation


ans =cor([1,2,3,4,4,8,9,10],[10,9,8,7,5,2,1,0])
print(ans)


```

```python
plt.scatter(data1,data2,label=f'correlation {ans:.2f}',color="red")
plt.xlabel("Data 1")
plt.ylabel("Data 2")
plt.title("Scatter Data 1 and Data 2")
plt.legend()
plt.show()
```
