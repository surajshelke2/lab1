```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sb

def sinplot(flip = 1):
 x=np.linspace(0, 14, 100)                            #setof hundred points
 for i in range(1, 5):
   plt.plot(x, np.sin(x + i * .5) * (7 - i) )
```

```python
sb.set_style({'axes.axisbelow':True,'grid.color' : 'yellow','axes.grid' : True,'axes.edgecolor' : 'white','axes.facecolor' : '#EAEAF2','axes.linewidth' : 9.0,'grid.linestyle' : u'-'})
sinplot()
sb.despine()
sb.set_palette('husl')                                        #remove top right border
print(plt.show())
```
