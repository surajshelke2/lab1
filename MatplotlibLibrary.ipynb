```python
#bar plot
x=['C','C++','Java','Python']
y=[90,95,98,91]
plt.xlabel('Languages',fontsize=15)
plt.ylabel('Marks',fontsize=15)
plt.title('MarksBar',fontsize=15)
c=['r','g','b','y']
#width-bar width,edgecolor-bar border color,linewidth-bar border width,linestyle-bar border style,alpha-color intensity
plt.bar(x,y,color=c,width=0.5,edgecolor='black',linewidth=3,linestyle=':',alpha=0.5)
plt.show()
```

```python
#bar plot
x=['C','C++','Java','Python']
y=[90,95,98,91]
z=[40,50,30,60]
plt.xlabel('Languages',fontsize=15)
plt.ylabel('Marks',fontsize=15)
plt.title('MarksBar',fontsize=15)
width=0.3;
plt.bar(x,y,width,color='r',label='LangBar')
plt.bar(x,z,width,color='b',label='LangBar')
plt.legend()
plt.show()
```

```python
#bar plot
import numpy as np
x=['C','C++','Java','Python']
y=[90,95,98,91]
z=[40,50,30,60]

width=0.3
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel('Languages',fontsize=15)
plt.ylabel('Marks',fontsize=15)
plt.title('MarksBar',fontsize=15)

plt.bar(p,y,width,color='r',label='LangBar')
plt.bar(p1,z,width,color='b',label='LangBar')

plt.xticks(p+width/2,x,rotation=25)
plt.legend()
plt.show()
```

```python
#bar plot
x=['C','C++','Java','Python']
y=[90,95,98,91]
z=[40,50,30,60]

plt.xlabel('Languages',fontsize=15)
plt.ylabel('Marks',fontsize=15)
plt.title('MarksBar',fontsize=15)

#barh for horizontal bar
plt.barh(x,y,color='r',label='LangBar')
plt.barh(x,z,color='b',label='LangBar')

plt.legend()
plt.show()
```

```python
countries = ['USA', 'India', 'China', 'Russia', 'Germany']
bronzes = np.array([38, 17, 26, 19, 15])
silvers = np.array([37, 23, 18, 18, 10])
golds = np.array([46, 27, 26, 19, 17])
ind = [x for x, _ in enumerate(countries)]

plt.bar(ind, golds, width=0.5, label='golds', color='gold', bottom=silvers+bronzes)
plt.bar(ind, silvers, width=0.5, label='silvers', color='silver', bottom=bronzes)
plt.bar(ind, bronzes, width=0.5, label='bronzes', color='#CD853F')

plt.xticks(ind,countries)
plt.ylabel("Medals")
plt.xlabel("Countries")
plt.legend(loc="upper right")
plt.title("2019 Olympics Top Scorers")
```
