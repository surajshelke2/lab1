```python
import numpy as np

data = np.linspace(1,50,200)
```

```python
mean =np.mean(data)
std =np.std(data)

def pdf(mean, std, data):
    probability = (1 / (np.sqrt(2 * np.pi) * std)) * np.exp(-0.5 * ((data - mean) / std) ** 2)
    return probability

x_pdf = pdf(mean,std,data)

print("PDF :",x_pdf[:5])
```

```python
plt.plot(data, x_pdf, label='PDF')
plt.title('Probability Density Function (PDF) of a Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```
